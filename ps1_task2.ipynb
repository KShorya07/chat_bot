```python
!pip install PyMuPDF
from tqdm.auto import tqdm
import fitz

def text_transformer(text:str)->str:
  cleaned_text=text.replace('\n' ,' ').strip()
  return cleaned_text
def open_and_read_file(pdf_path:str):

    pages_and_text=[]
    doc=fitz.open(pdf_path)
    for page_number,page in tqdm(enumerate(doc)):

      text=page.get_text()
      text=text_transformer(text)
      pages_and_text.append({"page_number": page_number,
                                "page_char_count": len(text),
                                "page_word_count": len(text.split(" ")),
                                "page_sentence_count_raw": len(text.split(". ")),
                                "page_token_count": len(text) / 4,
                                "text": text})
    return pages_and_text
pages_and_text=open_and_read_file(filename)
pages_and_text[:2]
```

```python
num_of_chunks=6
def chunking(input_list: list,
             slice_size: int)-> list[list[str]]:
 return [input_list[i:i+slice_size] for i in range(0, len(input_list), slice_size)]
for item in tqdm(pages_and_text):
  item['sentence_chunks']=chunking(input_list=item['sentence'],slice_size=num_of_chunks)
  item['num_of_chunks']=len(item['sentence_chunks'])
```

```python
for item in tqdm(pages_and_text):
  item['chunk_token_counts']=[]
  for chunk in item['sentence_chunks']:
    chunk_text=' '.join(chunk)
    chunk_token_count=len(chunk_text)/4
    item['chunk_token_counts'].append(chunk_token_count)
```

```python
import re
pages_and_chunk=[]
for item in tqdm(pages_and_text):
  for sentence_chunk in item['sentence_chunks']:
    chunk_dict={}
    chunk_dict['page_number']=item['page_number']
    joined_sentence_chunk="".join(sentence_chunk).replace('  ',' ').strip()
    joined_sentence_chunk=re.sub(r'\.(A-Z)',r'.\1',joined_sentence_chunk)
    chunk_dict['sentence_chunk']=joined_sentence_chunk
    chunk_dict['chunk_char_count']=len(joined_sentence_chunk)
    chunk_dict['chunk_word_count']=len([word for word in joined_sentence_chunk.split(' ')])
    chunk_dict['chunk_token_count']=len(joined_sentence_chunk)/4

    pages_and_chunk.append(chunk_dict)
len(pages_and_chunk)
```

```python
from sentence_transformers import SentenceTransformer
embedding_model=SentenceTransformer(model_name_or_path='all-mpnet-base-v2',
                                    device='cpu')

embedding=embedding_model.encode(pages_and_chunk[0]['sentence_chunk'],batch_size=32,
                                 convert_to_tensor=True)
embedding_dict=dict(zip(pages_and_chunk[0]['sentence_chunk'],embedding))

for sentence, emebdding in embedding_dict.items():
 print('sentence', sentence)
 print('embedding', embedding )
```

```python
import torch
from sentence_transformers import util
query='Why Tesla (TSLA) Skyrocketed During the Pandemic'
print(f'query {query}')
query_embedding=embedding_model.encode(query,convert_to_tensor=True)
from time import perf_counter as timer
start_time=timer()
dot_scores=util.dot_score(query_embedding,text_chunk_embedding).squeeze()#or [0]
end_time=timer()
top_results=torch.topk(dot_scores,k=5)
top_results
```

```python
def build_prompt(query, context_items=context_item):

    context = '\n\n'.join(context_items)
    prompt = f'''
You are a financial analysis assistant.
You are knowledgeable, professional, and explain financial topics in a clear and engaging way.

Start your answer with one of these phrases:
- "Absolutely, this is a critical point to consider."
- "Great question! Let’s explore this carefully."
- "Indeed, this topic is highly relevant to investors right now."

USE ONLY the information provided in the context.
Do NOT repeat the context word for word.
Explain clearly and concisely.
Use examples, analogies, or comparisons to make your answer more engaging.
Highlight key points in bullet points when relevant.
Provide a short takeaway or insight at the end for better understanding.

Question:
{query}

Context:
{context}

Answer:
'''
    return prompt
```

```python
def ask(query):
      prompt=build_prompt(query)
      response=client.chat.completions.create(
      model='llama-3.1-8b-instant',
      messages=[{'role':'user', 'content':prompt}],
      temperature=.2)

      print(f'query: \n {query}')
      print(f'answer')
      output=response.choices[0].message.content
      print_wrapped(output)
```
